```python
#Hypothesis testing
import numpy as np
from scipy import stats
```

```python
#Two groups of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
```

```python
#Perform t_test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
print(t_stat)
print(p_value)
```

```python
alpha=0.05
```

```python
if p_value<alpha:
  print("Reject the null hypothesis. The means are significant difference")
else:
  print("Fail to reject alternative hypoythesis .The means are not significant difference")
```

```python
import pandas as pd
df=pd.read_csv("/content/Vaccine.csv")
df
```

```python
a=df['Efficiency'].mean()
b=df['Rec_Rate'].mean()
print("Efficiency mean=",a)
print("Recovery rate=",b)
```

```python
group_A=df['Efficiency']
group_B=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(group_A,group_B,equal_var=False)
```

```python
alpha=0.05
```

```python
if p_value>alpha:
  print("H0-related")
else:
  print("H1-not related")
```

ANOVA

```python
dfa=df[df['Vaccine']=="A"]['Efficiency']
dfb=df[df['Vaccine']=="B"]['Efficiency']
dfa
```

```python
f_stat,p_value=stats.f_oneway(dfa,dfb)
print("F_statistic:",f_stat)
print("p_value:",p_value)
```

```python
alpha=0.05
```

```python
if p_value>alpha:
  print("H0-related")
else:
  print("H1-not related")
```
