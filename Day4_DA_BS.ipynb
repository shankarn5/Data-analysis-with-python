extracting HTML code of a website

program flow :

1.importing libraries

2.save the url

3.using requests.get(),access the webpage

4.using BeautifulSoup(), access the HTML code

```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rq
```

```python
url="https://www.geeksforgeeks.org"
```

```python
page=rq.get(url)
```

```python
hc=bs(page.text,'html')
hc
```

```python
title=hc.find('title')#To get title of a website
title
```

```python
print(title.text)#removes the tags of an html code
```

Extracting an element in the webpage

```python
url1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page1=rq.get(url1)
hc1=bs(page1.text,'html')
hc1
```

```python
t=hc1.find('table')
print(t.text)

```

```python
h=t.find_all('th')
h
```

```python
headings=[i.text for i in h]
headings
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=headings)
print(mydf)
```

```python
rows=t.find_all('tr')
print(rows)
```

```python
rowdata=t.find_all('td')
data=[i.text for i in rowdata]
data



```

```python

print(data[0:4])


```

```python
st=0
sp=4
for i in range((len(data)//4)):



  mydf.loc[i]=data[st:sp]
  st=st+4
  sp=sp+4
mydf

```

```python
mydf.to_csv("dfrmi.csv")
```
